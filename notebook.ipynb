```python
from sklearn.preprocessing import LabelEncoder,StandardScaler,PolynomialFeatures,OneHotEncoder
from sklearn.model_selection import train_test_split,KFold,GridSearchCV
from sklearn.compose import ColumnTransformer
from xgboost import XGBClassifier
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.pipeline import Pipeline
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense

#prepare x and y
y = df['PlayerType']
x = df.drop(columns=['PlayerType'])

#label encode the target (y)
le = LabelEncoder()
y_encoded = le.fit_transform(y)

#train-test split on raw data - this prevents leakage
x_train,y_train,x_test,y_test = train_test_split(
    x,y_encoded,test_size=0.2,random_state=42,stratify=y_encoded
)

#identify numerical and categorical column
num_cols = x_train.select_dtypes(include='number').columns
cat_cols = x_train.select_dtypes(include='object').columns

preprocessor = ColumnTransformer(transformers=[
    ('scaler',StandardScaler(),num_cols),
    ('poly',PolynomialFeatures(include_bias=False),num_cols)
    ('onehot',OneHotEncoder(handle_unknown='ignore'),cat_cols)
])

#Multiclass xgboost with probabilities
xgb = XGBClassifier(
    objective='multi:softprob',
    num_classes = len(np.unique(y_encoded)),
    use_label_encoder = False,
    eval_metric = 'mlogloss',
    random_state = 42,
    verbosity = 0,
    n_jobs = -1
)

pipe_xgb = Pipeline(steps=[
    ('preprocessor',preprocessor),
    ('model',xgb)
])


cv = KFold(n_splits=5,shuffle=True,random_state=42)

model = XGBClassifier(objective='multi:softmax',verbosity=0,random_state=42,scale_pos_weight=True)


```
